```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/catWind.jpg")
cv2_imshow(image)
```

```python
print("shape:", image.shape)
# 1024 height
# 768 width
# 3 color channels
```

```python
(b, g, r) = image[100, 50]
print("blue", b, "green", g, "red", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray, (400, 350))
cv2_imshow(resized)
```

```python
cropped = gray [200:300, 600:700]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("crappedCat.jpg", flipped)
```
